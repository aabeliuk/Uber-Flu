```python
import pandas as pd
import numpy as np
from datetime import datetime
mapping = dict(zip(cities, ["Chicago", "Houston", "Philadelphia", "Boston", "NewYork", "SanFrancisco", "Phoenix", "LosAngeles", "Miami", "Seattle"]))
city_min_temp_df = pd.DataFrame(columns=["Chicago", "Houston", "Philadelphia", "Boston", "NewYork", "SanFrancisco", "Phoenix", "LosAngeles", "Miami", "Seattle"])
city_min_temp_dict = {}
city_max_temp_df = pd.DataFrame(columns=["Chicago", "Houston", "Philadelphia", "Boston", "NewYork", "SanFrancisco", "Phoenix", "LosAngeles", "Miami", "Seattle"])
city_max_temp_dict = {}
city_prcp_dict = {}
city_snow_dict = {}
#temp being 10ths of degree C, prcp 10ths of mm, snow mm
```

```python
#for city in cities:
for city in cities:
    city_min_temp_dict[mapping[city]] = {}
    city_max_temp_dict[mapping[city]] = {}
    city_prcp_dict[mapping[city]] = {}
    city_snow_dict[mapping[city]] = {}
    for year in range(2004,2018):
        print(city, year)
        offset=1
        count = 1000000
        
        while offset < count:
            r = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&locationid={}&startdate={}-01-01&enddate={}-01-01&limit=1000&datatypeid=TMAX&datatypeid=TMIN&datatypeid=SNOW&datatypeid=PRCP&offset={}'.format(city, year, year+1, offset), headers=headers)
            try:
                json = r.json()
            except:
                print(r.text)
                raise Exception
            offset = json['metadata']['resultset']['offset']
            if offset == 1:
                count = json['metadata']['resultset']['count']
            print(offset)            

            for x in json['results']:

                if x['datatype'] == 'TMAX':
                    city_temp_dict = city_max_temp_dict
                elif x['datatype'] == 'TMIN':
                    city_temp_dict = city_min_temp_dict
                elif x['datatype'] == 'SNOW':
                    city_temp_dict = city_snow_dict
                elif x['datatype'] == 'PRCP':
                    city_temp_dict = city_prcp_dict
                cur_date = datetime.strptime(x['date'], "%Y-%m-%dT00:00:00")
                cur_mo = "{}-{}-{}".format(cur_date.month, cur_date.day - (cur_date.weekday() + 1), cur_date.year)
                if cur_mo not in city_temp_dict[mapping[city]]:

                    val = str(x['value'])
                    val = "{}.{}".format(val[:-1], val[-1])
                    city_temp_dict[mapping[city]][cur_mo] = [float(val)]
                else:
                    val = str(x['value'])
                    val = "{}.{}".format(val[:-1], val[-1])
                    city_temp_dict[mapping[city]][cur_mo] += [float(val)]
            offset += 1000
        for city_temp_dict in [city_min_temp_dict, city_max_temp_dict, city_prcp_dict, city_snow_dict]:
            for date in city_temp_dict[mapping[city]]:
                city_temp_dict[mapping[city]][date] = np.mean(city_temp_dict[mapping[city]][date])
    

        
```
